# Exploratory Data Analysis and Visualization of Covid-19 Data Using Python

### Importing Libraries

```python
# Import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
# Set style & figures inline
sns.set()
%matplotlib inline
```

### Loading Data

```python
confirmed_cases_data = 'time_series_covid19_confirmed_global.csv'
death_cases_data = 'time_series_covid19_deaths_global.csv'
recovered_cases_data = 'time_series_covid19_recovered_global.csv'

# Import datasets as pandas dataframes
raw_data_confirmed = pd.read_csv(confirmed_cases_data)
raw_data_deaths = pd.read_csv(death_cases_data)
raw_data_recovered = pd.read_csv(recovered_cases_data)
```

```python
# Function ubah format tanggal
def convert_date(df):
  df.columns = list(df.columns[:4]) + [datetime.strptime(d, "%m/%d/%y").date().strftime("%m-%d-%y") for d in df.columns[4:]]
```

```python
# Copy dataframe raw_data_confirmed ke df confirmed_country
confirmed_country = raw_data_confirmed.copy()

# Panggil function untuk ubah format tanggal
convert_date(confirmed_country)

# Satukan baris yang terpisah provinsi ke satu negara yang sama
confirmed_country = confirmed_country.groupby(['Country/Region']).sum()

# Hapus kolom 'Lat', dan 'Long' karena tidak diperlukan
confirmed_country = confirmed_country.drop(columns=['Lat', 'Long'])

# Ubah kolom jadi baris dan baris jadi kolom
confirmed_country = confirmed_country.transpose()

# Jadikan tanggal ('Date') sebagai index dari DataFrame
confirmed_country.index.names = ['Date']

confirmed_country.head()
```

Berdasarkan hasil diatas, dapat dilihat pada tanggal 22 Januari 2020 terdapat 548 kasus terkonfirmasi positif Covid-19 di negara China. Hal tersebut menunjukkan bahwa China lah sumber awal penyebaran virus Covid-19. Selain itu, pada tanggal yang sama dapat dilihat juga pada negara-negara pada Benua Asia sekitar China seperti Jepang, Korea Selatan, Taiwan, dan Thailand juga terdapat kasus terkonfirmasi positif Covid-19. Selain kawasan Asia, pada tanggal yang sama di Amerika Serikat juga terdapat 1 kasus terkonfirmasi Covid-19. Hal tersebut menunjukkan pada tanggal 22 Januari 2020 tersebut, penyebaran Covid-19 sudah keluar dari Benua Asia sampai ke Benua Amerika namun belum sampai Benua lainnya.

### Confirmed Cases Growth in 6 Countries

```python
# Slice DataFrame untuk mengambil 6 negara yang diminta
confirmed_six = confirmed_country[['China', 'Spain', 'France', 'Italy', 'US', 'Australia']].reset_index()

# Ubah bentuk DataFrame -> negara-negara dijadikan dalam 1 kolom 'Country'
confirmed_six = confirmed_six.melt('Date', var_name='Country', value_name='Cases')
```

```python
# Visualisasi Data 6 negara berdasarkan waktu

sns.set_style('whitegrid')
fig, ax = plt.subplots(figsize=(10, 5))

confirmed_six_plot = sns.lineplot(x='Date', y='Cases', hue='Country',  data=confirmed_six, ax=ax, linewidth=3)

confirmed_six_plot.set_title('Confirmed Cases Growth in Australia, China, France, Italy, Spain, and USA', weight='bold', fontsize=18)
confirmed_six_plot.set_xlabel('Date', weight='bold', fontsize=15)
confirmed_six_plot.set_ylabel('Cases', weight='bold', fontsize=15)
confirmed_six_plot.set_xticks(confirmed_six_plot.get_xticks()[::10])

plt.setp(confirmed_six_plot.get_legend().get_texts(), fontsize='10') 
plt.setp(confirmed_six_plot.get_legend().get_title(), fontsize='15') 

plt.show()
```

Berdasarkan hasil visualisasi di atas, dapat dilihat di negara Australia penyebaran Covid-19 adalah yang terendah dibandingkan kelima negara lainnya. Dapat dilihat juga walaupun China sempat mengalami pelonjakan kasus terkonfirmasi Covid-19 lebih dahulu dari kelima negara lainnya, namun mulai dari sekitar pertengahan Februari kasus terkonfirmasi di China cenderung stabil. Di negara-negara Benua Eropa seperti Italia, Spanyol, dan Perancis mulai dari bulan Maret kasus terkonfirmasi covid-19 mulai meningkat secara signifikan bahkan sampai melebihi total terkonfirmasi Covid-19 di China pada bulan April awal. Dapat dilihat juga di Amerika Serikat, walaupun dari bulan Januari hingga Maret kasus terkonfirmasi cenderug stabil, namun mulai dari pertengahan Maret dapat dilihat terjadi pelonjakan jumlah kasus terkonfirmasi yang sangat signifikan. Di Amerika Serikat juga sampai melewati jumlah kasus terkonfirmasi di China bahkan sebelum bulan April dan di awal April menjadi negara dengan jumlah kasus terkonfirmasi terbanyak diantara enam negara tersebut.

```python
# Copy dataframe raw_data_deaths ke variable deaths
deaths = raw_data_deaths.copy()

# Panggil function untuk ubah format tanggal
convert_date(deaths)

# Satukan baris yang terpisah provinsi ke satu negara yang sama
deaths = deaths.groupby(['Country/Region']).sum()

# Hapus kolom 'Lat', dan 'Long' karena tidak diperlukan
deaths = deaths.drop(columns=['Lat', 'Long'])

# Tukar kolom jadi baris dan baris jadi kolom
deaths = deaths.transpose()

# Jadikan tanggal ('Date') sebagai index dari DataFrame
deaths.index.names = ['Date']

deaths.tail()
```

```python
# Cek indeks

deaths.index
```

Berdasarkan data di atas, dapat dilihat walaupun China adalah negara asal mula penyebaran Covid-19 ditambah China sebagai peringkat pertama negara dengan populasi dunia terbanyak, namun jumlah kematian akibat Covid-19 sampai dengan 7 April 2020 di China dapat terhitung rendah jika dibandingkan negara-negara lainnya yaitu sebanyak 3.335 orang. Jumlah kematian penduduk di negara-negara lain sampai dengan 7 April 2020, di Britania Raya sebanyak 6.171 orang, Prancis sebanyak 10.343 orang, Amerika Serikat sebanyak 12.722 orang, dan Spanyol sebanyak 14.045 orang, Italia sebanyak 17.127 orang.

```python
deaths_25 = deaths.reset_index()

# Ubah bentuk DataFrame -> negara-negara digabung dalam 1 kolom 'Country' dengan valuenya kolom 'Deaths'
deaths_25 = deaths_25.melt('Date', var_name='Country', value_name='Deaths')

# Seleksi DataFrame dengan kematian setidaknya 25 orang
deaths_25 = deaths_25.loc[deaths_25['Deaths'] >= 25]

# Urutkan DataFrame berdasarkan tanggal
deaths_25 = deaths_25.sort_values('Date')

deaths_25.head()
```

Berdasarkan hasil di atas, dapat dilihat negara China pada tanggal 24 Januari 2020 sudah mencapai kasus kematian lebih dari 25 orang yaitu sebanyak 26 orang. Lalu negara selain China yang pertama kali menembus jumlah kematian 25 adalah negara Iran pada tanggal 27 Februari yang menyentuh angka 26 orang. Disusul negara Italia yang menembus angka 25 pada tanggal 29 Februari sebanyak 29 orang. Dari ketiga negara tersebut, dua negara berasal dari Asia. Hal tersebut kasus kematian yang melebihi 25 orang baru terjadi di kawasan Asia.

```python
# Seluruh negara dijadikan satu untuk menampilkan data kesembuhan di dunia
deaths_25 = deaths_25.groupby('Date').sum()

# Visualisasi Heat Map
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(deaths_25, ax=ax)
ax.set_title('Heat Map Growth Death in The World', weight='bold', fontsize=18)
```

Dari hasil visualisasi di atas, dapat dilihat secara keseluruhan dunia kasus kematian akibat Covid-19 selama kurang dari 3 bulan cukup lambat mencapai 40.000 kasus. Namun, yang mengejutkan dalam waktu beberapa hari saja, di bulan April terjadi kelonjakan sebanyak 40.000 kasus kematian sehingga total kasus mencapai lebih dari 80.000 kasus. Sehingga dapat dibilang bahwa kelonjakan kasus kematian di seluruh dunia terjadi mulai dari bulan April.

### Recovered Patient From Covid-19

```python
# Function

def convert_df(raw_df):
  df = raw_df.copy()
  convert_date(df)
  df = df.groupby(['Country/Region']).sum()
  df = df.drop(columns=['Lat', 'Long'])
  df = df.transpose()
  df.index.names = ['Date']
  return df
```

```python
recovered = convert_df(raw_data_recovered)
recovered = recovered.iloc[:77]
recovered.head()
```

```python
recovered_world = recovered.reset_index()

recovered_world = recovered_world.melt('Date', var_name='Country', value_name='Recovered Patient')
```

```python
# Seluruh negara dijadikan satu untuk menampilkan data kesembuhan di dunia
recovered_world = recovered_world.groupby('Date').sum()

# Visualisasi Heat Map
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(recovered_world, ax=ax)
ax.set_title('Heat Map Pasien Covid-19 Sembuh di Dunia', weight='bold', fontsize=18)
```

Berdasarkan Heat Map di atas, dapat dilihat total kesembuhan pasien Covid-19 di dunia cukup lambat. Kesembuhan pasien sampai dengan bulan Februari yaitu sebanyak 50.000 orang yang hanya disumbangkan oleh negara China. Sedangkan untuk kebanyakan negara, mulai dari bulan Maret baru terjadi peningkatan yang signifikan untuk jumlah pasien yang sembuh. Sehingga kesembuhan pasien dunia mencapai lebih dari 100.000 mulai dari bulan Maret dan mencapai kesembuhan lebih dari 200.000 orang mulai dari bulan April.

```python
# Slice DataFrame untuk mengambil 6 negara yang diminta
recovered_six = recovered[['China', 'Spain', 'France', 'Italy', 'US', 'Australia']].reset_index()

# Ubah bentuk DataFrame -> negara-negara dijadikan dalam 1 kolom 'Country'
recovered_six = recovered_six.melt('Date', var_name='Country', value_name='Recovered Patient')
```

```python
# Visualisasi Data Pasien Sembuh di 6 negara berdasarkan waktu

sns.set_style('whitegrid')
fig, ax = plt.subplots(figsize=(10, 5))

recovered_six_plot = sns.lineplot(x='Date', y='Recovered Patient', hue='Country',  data=recovered_six, ax=ax, linewidth=3)

recovered_six_plot.set_title('Recovered Patient in Australia, China, France, Italy, Spain, and USA', weight='bold', fontsize=18)
recovered_six_plot.set_xlabel('Date', weight='bold', fontsize=15)
recovered_six_plot.set_ylabel('Recovered Patient', weight='bold', fontsize=15)
recovered_six_plot.set_xticks(recovered_six_plot.get_xticks()[::10])

plt.setp(recovered_six_plot.get_legend().get_texts(), fontsize='12') 
plt.setp(recovered_six_plot.get_legend().get_title(), fontsize='15') 

plt.show()
```

Berdasarkan hasil visualisasi di atas, dapat dilihat diantara keenam negara, China sudah dapat mengobati pasien Covid-19 paling cepat dan yang kenaikan sembuhnya pasien paling tinggi dibandingkan negara lainnya. Lalu pada awal April, China juga yang tertinggi dalam pasien sembuh Covid-19. Dapat dilihat pula, negara Spanyol memiliki tingkat kesembuhan pasien Covid-19 tertinggi diantara negara Eropa lainnya seperti Italia dan Prancis. Sedangkan untuk negara Amerika Serikat, tingkat kesembuhan pasien sangat rendah dibandingkan dengan kurva terkonfirmasi dan meninggalnya. Untuk negara Australia, tingkat kesembuhannya cenderung rendah dan setara dengan kurva terkonfirmasi dan meninggalnya.
